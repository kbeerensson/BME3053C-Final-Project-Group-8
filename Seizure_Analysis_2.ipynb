```python
import os
import numpy as np
import mne
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt

# 1. File labeling based on what you know about the seizure content
seizure_files = ['chb01_03.edf', 'chb02_16.edf']
non_seizure_files = ['chb01_04.edf', 'chb02_19.edf']
all_files = seizure_files + non_seizure_files

# 2. Folder containing all the EDF files in your GitHub repo
data_folder = "data"  # <- this is now the GitHub 'data/' folder

# 3. Function to load data from files
def load_data_from_files(folder, file_list, label, duration=10, overlap=5, max_epochs=50):
    X, y = [], []
    for file in file_list:
        file_path = os.path.join(folder, file)
        raw = mne.io.read_raw_edf(file_path, preload=True, verbose=False)
        raw.filter(0.5, 40, fir_design='firwin')

        # Epoch into 10s windows with 5s overlap
        epochs = mne.make_fixed_length_epochs(raw, duration=duration, overlap=overlap, preload=True)
        data = epochs.get_data()[:max_epochs]  # limit number of epochs

        X.append(data)
        y.extend([label] * data.shape[0])

    return np.concatenate(X), np.array(y)

# 4. Load and label data
X_seizure, y_seizure = load_data_from_files(data_folder, seizure_files, label=1)
X_non_seizure, y_non_seizure = load_data_from_files(data_folder, non_seizure_files, label=0)

# 5. Combine, scale, and reshape data
X = np.concatenate([X_seizure, X_non_seizure])
y = np.concatenate([y_seizure, y_non_seizure])

# Normalize each epoch (use only the first channel for simplicity)
scaler = StandardScaler()
X_scaled = np.array([scaler.fit_transform(epoch.T).T for epoch in X])  # shape: (samples, timepoints)
X_scaled = X_scaled[..., np.newaxis]  # Add channel dimension for CNN

# 6. Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 7. Build a basic CNN
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(X_train.shape[1], X_train.shape[2], 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# 8. Train the model
history = model.fit(X_train, y_train, epochs=10, batch_size=8, validation_split=0.2)

# 9. Evaluate the model
y_pred = (model.predict(X_test) > 0.5).astype("int32")
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

# 10. Plot accuracy & loss
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Val')
plt.title('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Val')
plt.title('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# === 1. Load the EDF file ===
file_path = 'data/chb01_03.edf'  # Make sure this matches your folder & filename
raw = mne.io.read_raw_edf(file_path, preload=True)

# === 2. Basic Info (Optional) ===
print("Channels:", raw.ch_names)
print("Sampling rate:", raw.info['sfreq'], "Hz")
print("Recording duration:", raw.times[-1] / 60, "minutes")

# === 3. Plot interactive EEG viewer ===
raw.plot(n_channels=10, duration=30, scalings='auto', title='Raw EEG Signal')
```
